### cleaning

```python
print(f'duplicate values = {df.duplicated().sum()}')
print(f'missing values = {df.isnull().sum().sum()}')
```

```python
df['last_updated'] = df['last_updated'].astype('datetime64[ns]')
```

```python

# df[['sunrise', 'sunset', 'moonrise', 'moonset']] = pd.to_datetime(df[['sunrise', 'sunset', 'moonrise', 'moonset']])
# handling AM/PM format to minutes and no moonrise/moonset values


def to_minutes(x):
    if isinstance(x, str) and ("AM" in x or "PM" in x):
        t = pd.to_datetime(x).time()
        return t.hour * 60 + t.minute
    return None 

column = ['sunrise', 'sunset', 'moonrise', 'moonset']

for i in column:
    df[i] = df[i].apply(to_minutes)
```
